# Mosques  Data

```python
import pandas as pd
import json
import csv
import time
import sys
import http.client, urllib.request, urllib.parse, urllib.error, base64
import googletrans 
from googletrans import Translator
```

## Data Request

```python
astrings1 = list()


try:
    conn = http.client.HTTPSConnection('apina.address.gov.sa')
    for page_num in range(1, 198):
                    conn.request("GET", "https://apina.address.gov.sa/NationalAddress/v3.1/address/poi-fixed-params?language=E&format=JSON&ServiceString=Mosque&cityname=riyadh&api_key=e5f9ea9216e446be97227cf99e679618&page=" + str(page_num))
                    response = conn.getresponse()
                    data= response.read()
                    decode=data.decode('utf-8', 'ignore')
                    astrings1.append(decode)
                    time.sleep(10)
                    conn.close()                
except Exception as e:
    print("[Errno {0}] {1}".format(e.errno, e.strerror))

####################################
```

## Data Assesment

```python
df=pd.Series(astrings1).to_frame()
```

```python
df.tail(60)
```

## Create a Dataframe

```python
x=0 
for index, row in df.iterrows():
    if x==0:
              y=df[0][index]
              result=pd.read_json (y)
    else:
                y=df[0][index]
                if "Internal server error" in y :
                    print()
                else:
                    result=result.append(pd.read_json (y))
    x=x+1 
```
